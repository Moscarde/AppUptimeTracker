```python
from datetime import datetime
import pandas as pd
```

```python
date = '2023-10-29'
file = open(f"{date}.txt", "r", encoding="UTF-8")
data = file.read().split("\n")
```

```python
data
```

```python
separator_index = data.index('<~~~~~~~~~~~~~~~~~~~~~~~~ Usage time - Specific ~~~~~~~~~~~~~~~~~~~~~~~~>')
print(separator_index)
data_sum = data[1:separator_index - 1]
data_log = data[separator_index + 1:]
print(data_sum)
print(data_log)
```

```python
dicionario = {'total_time': [], 'app_name': [],}
for log in data_sum:
    time, app_name = list(log.split('  <~~>  '))

    dicionario['app_name'].append(app_name)
    time = datetime.strptime(time, '%H:%M:%S').time()
    dicionario['total_time'].append(time)

df = pd.DataFrame(dicionario)
df = df.sort_values(by='total_time', ascending= False)
df.to_csv(f'{date} - SUM.csv', index=False)
df
```

```python
dicionario = {'total_time': [], 'app_name': [],}
for log in data_log:
    if log == '':
        continue
    time, app_name = list(log.split('  <~~>  '))

    dicionario['app_name'].append(app_name)
    time = datetime.strptime(time, '%H:%M:%S').time()
    dicionario['total_time'].append(time)

df = pd.DataFrame(dicionario)
df = df.sort_values(by='total_time', ascending= False)
df.to_csv(f'{date} - LOG.csv', index=False)
df
```
